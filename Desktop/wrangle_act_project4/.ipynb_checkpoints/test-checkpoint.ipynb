## Cleaning data: 
Here, I will do the data wrangling based on process:
- Define 
- Code
- Test 





```python
# Copy dataFarme:
TA2= TA.copy()
IP2 = IP.copy()
J2 = J.copy()
```

#### Define: 
- The three data sets need to be merged 

#### Code

```python
# merge TA2 & J2 by tweet_id .
M1= pd.merge(TA2,J2, left_on = 'tweet_id', right_on = 'tweet_id', how = 'inner')

# merge M1 with IP2 (Image Predictions)
M2= pd.merge(M1, IP2, left_on = 'tweet_id', right_on = 'tweet_id', how = 'inner')

M2
```

#### Test

```python
M2.info()
```

```python
TA2.info()
```

#### Define
- The Sources should be clear and can be read them. 

#### Code

```python
M2['source_x'] = M2['source_x'].apply(lambda x: re.findall(r'>(.*)<', x)[0])  
```

#### Test

```python
M2['source_x'].value_counts()
```

#### Define
- Here, combined these columns P1 ,P2 and P3 and cleaned them. 

#### Code

```python
prediction = []
confidence = []

def get_prediction_confidence(dataframe):
    if dataframe['p1_dog'] == True:
        prediction.append(dataframe['p1'])
        confidence.append(dataframe['p1_conf'])
    elif dataframe['p2_dog'] == True:
        prediction.append(dataframe['p2'])
        confidence.append(dataframe['p2_conf'])
    elif dataframe['p3_dog'] == True:
        prediction.append(dataframe['p3'])
        confidence.append(dataframe['p3_conf'])
    else:
        prediction.append('NaN')
        confidence.append(0)

M2.apply(get_prediction_confidence, axis=1)
M2['prediction'] = prediction
M2['confidence'] = confidence
```

```python
# drop the not need columns 
M2 = M2.drop(['img_num', 'p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog'], 1)
```

#### Test

```python
M2.info()
```

#### Define
- Here, Create one column for Dog life stage. 

#### Code

```python
# Select melt and stay columns
MELTS_COLUMNS = ['doggo', 'floofer', 'pupper', 'puppo']
STAY_COLUMNS = [x for x in M2.columns.tolist() if x not in MELTS_COLUMNS]

# Melt the columns
M2 = pd.melt(M2, id_vars = STAY_COLUMNS, value_vars = MELTS_COLUMNS, 
                         var_name = 'stages', value_name = 'dog_stage')
# Drop superfluous column
M2 =M2.drop('stages', 1)
```

#### Test

```python
print('Value Counts' + '\n')
print(M2.dog_stage.value_counts())
```

```python
M2.info()
```

#### Define
- Removing rows that have non-empty retweeted_status_id, retweeted_status_user_id, and retweeted_status_timestamp



#### Code

```python
M2 = M2[(M2['retweeted_status_timestamp'].isna() == False) | 
                (M2['retweeted_status_id'].isna() == False) |
    (M2['retweeted_status_user_id'].isna() == False)]

M2.reset_index(inplace=True, drop=True)

M2.reset_index(inplace=True, drop=True)
```

#### Test

```python
M2.head()  
```

#### Define
- Clean and dropping duplicates row and not need columns.


#### Code

```python
M2 = M2[pd.isnull(M2.retweeted_status_id)]

# Delete columns we don't need 
M2 = M2.drop('retweeted_status_id', 1)
M2 = M2.drop('retweeted_status_user_id', 1)
M2 = M2.drop('retweeted_status_timestamp', 1)

# Delete duplicated tweet_id
M2= M2.drop_duplicates()

# Delete tweets with no pictures
M2 = M2.dropna(subset = ['jpg_url'])

#Delete dog_stage duplicates
M2 = M2.sort_values('dog_stage').drop_duplicates('tweet_id', keep = 'last')
```

#### Test

```python
M2.info()
```

```python
M2.dog_stage.value_counts() 
```

#### Define

- Rating_denominator has minimum value as 0.

#### Code

```python
null_index = M2[M2['rating_denominator'] == 0].index
M2 = M2.drop(null_index)
```

#### Test

```python
M2[M2['rating_denominator'] == 0]
```

#### Define
- Change the values of 'NaN' to type of 'None' 

#### Code

```python
M2.loc[M2['prediction'] == 'NaN', 'prediction'] = None
M2.loc[M2['confidence'] == 'NaN', 'confidence'] = None

M2.loc[M2['rating_numerator'] == 'NaN', 'rating_numerator'] = 0
```

#### Test

```python
M2.info()
```

#### Define
- Here, Correct datatypes.

#### Code

```python
M2['tweet_id'] = M2['tweet_id'].astype(str)
M2['timestamp'] = pd.to_datetime(M2.timestamp)
M2['source'] = M2['source_x'].astype('category')
M2['source'] = M2['source_y'].astype('category')
M2['favorited'] = M2['favorited'].astype(int)
M2['retweeted'] = M2['retweeted'].astype(int)
M2['dog_stage'] = M2['dog_stage'].astype('category')
```

#### Test 

```python
M2.dtypes
```

#### Define
- There are some retweets that we can point out as they start with 'RT' in the text column.

#### Code

```python
retweets = M2.text.str.match('^RT @')
M2 = M2[~retweets]
```
